```python
# Dependencies
import pandas as pd

```

```python
# load CSV
coders = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
coders_df = pd.read_csv(coders, low_memory=False)
coders_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_df = coders_df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_df.head()
```

```python
list(coders_df)
```

```python
# Change reading 0 and 1 to No and Yes, respectively
reduced_df= reduced_df.replace({0.0: "No", 1.0: "yes"})
reduced_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended = reduced_df.loc[reduced_df["AttendedBootcamp"] == "yes"]
attended.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
dif_camps = pd.DataFrame(coders_df["BootcampName"].value_counts())
dif_camps.reset_index(inplace=True)
dif_camps.columns = ["BootcampName", "Count"]
dif_camps.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp 

bootcamp_recommend = attended.replace({"Yes": 1, "No": 0})
bootcamp_recommend = pd.DataFrame(dif_camps.groupby("BootcampName")["Count"].sum())

bootcamp_recommend.reset_index(inplace=True)
bootcamp_recommend.columns = ["BootcampName", "Recommend"]

bootcamp_recommend.head ()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_table = pd.merge(bootcamp_recommend, dif_camps, on="BootcampName")
merge_table.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
percent_recommend = (merge_table["Recommend"]/merge_table["Count"]*100).map("{:,.2f}%".format)
percent_recommend.head()
```

```python
merge_table["percent_reccomend"]= merge_table["Recommend"]/\
merge_table['Count']
merge_table.head()
```

```python
merge_table['Recommend Rate'] = (merge_table['Recommend']/merge_table['Count']).map("{0:,.2f}%".format)
merge_table.head()
```
